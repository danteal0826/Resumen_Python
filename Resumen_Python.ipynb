#Librerías

Para poder inciar con un programa que requiera resolver numericamente un sistema de ecucaciones lineales, es necesario importar la biblioteca Numpy

Para poder graficar los datos que tengamos es necesario importar la librería matplotlib que es la librería con los recursos para generar graficas de datos numericos

```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigue crear las dos matrices con los sistemas de ecuaciones dados, en este caso:

3x+9y-10z=24
x-6y+4z=-4
10x-2y+8z=20

```python
# Matriz de coeficientes
A = np.array([[3, 9, -10],
              [1, -6, 4],
              [10, -2, 8]])

# Matriz de términos independientes
B = np.array([24, -4, 20])
```

Luego sigue resolver el sistema de ecuaciones con atributo que tiene la biblioteca numpy que es linalg.solve el cual necesita que se le entreguen dos matrices (A, B)

```python
sol = np.linalg.solve(A, B)
print("Las soluciones son:")
print("x =", sol[0])
print("y =", sol[1])
print("z =", sol[2])
```

Con el sistema de ecuaciones resuleto sigue la parte de graficar en donde se recurrira a la librería matplotlib anteriormente importada, como hay 3 variables indica que es un sistema en 3D, por lo tanto toca adicionar  submódulo de la biblioteca Matplotlib para poder graficar en 3D. mpl_toolkits.mplot3d y Axes3D es una clase dentro del submódulo

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Crear una malla de puntos en 3D
x_vals = np.linspace(-10, 10, 400) #genera 400 valores equidistantes entre -10 y 10 para X y Y
y_vals = np.linspace(-10, 10, 400) #Lo mismo para Y
X, Y = np.meshgrid(x_vals, y_vals) #toma estos valores y crea dos matrices 2D: X y Y. Cada matriz contiene todas las combinaciones posibles de X y Y
```

#Grid
es una estructura de datos que contiene todos los pares posibles de X y Y dentro de un rango específico

Sigue calcular los valores que corresponden a Z en cada una de las ecuaciones dadas, esto se hace para poder graficar cada una de las ecuaciones como superficies en un espacio tridimensional.


```python
# Ecuaciones de las superficies
Z1 = (24 - 3*X - 9*Y) / -10
Z2 = (-4 - X + 6*Y) / 4
Z3 = (20 - 10*X + 2*Y) / 8
```

Sigue graficar la figura en 3D
se le asigna a fig->plt.figure(): Esta función crea una nueva figura en la que se pueden dibujar gráficos

a Ax se le asigna la funcion add_subplot que puede crear gráficos 2D y 3D, dentro de esta va 111, projection='3d'. El 111 corresponde a la disposicion de subplot
El primer 1 indica el número de filas de subplots (1 en este caso).
El segundo 1 indica el número de columnas de subplots (1 en este caso).
El tercer 1 indica el índice del subplot actual (el primer subplot en este caso).
y el projection='3d' corresponde a que el subplot debe ser un gráfico 3D


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Graficar las superficies
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100, color='r', label='3x + 9y - 10z = 24')
ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100, color='g', label='x - 6y + 4z = -4')
ax.plot_surface(X, Y, Z3, alpha=0.5, rstride=100, cstride=100, color='b', label='10x - 2y + 8z = 20')

# Graficar el punto de intersección
ax.scatter(sol[0], sol[1], sol[2], color='k', s=100)

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la gráfica
plt.show()
```
